# 인물사진을 만들어 보자
아웃포커싱 인물사진, 아웃포커싱 동물 사진, 배경전환 크로마키사진을 한장씩 만들어 보자

### 1-1. 아웃포커싱 인물사진

```python
import os
import urllib
import cv2
import numpy as np
from pixellib.semantic import semantic_segmentation
from matplotlib import pyplot as plt
```

아웃포커싱을 적용할 인물 이미지를 출력해 보았다.

```python
img_path = os.getenv('HOME')+'/aiffel/human_segmentation/images/person_pic.png'  
img_orig = cv2.imread(img_path) 

print(img_orig.shape)

plt.imshow(cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB))
plt.show()
```

이미지 인식에 사용할 모델(PixLib에서 제공)을 urllib를 사용하여 가져와, model_file 이름 및 경로로 저장한다.

```python
model_dir = os.getenv('HOME')+'/aiffel/human_segmentation/models' 
model_file = os.path.join(model_dir, 'deeplabv3_xception_tf_dim_ordering_tf_kernels.h5') 
model_url = 'https://github.com/ayoolaolafenwa/PixelLib/releases/download/1.1/deeplabv3_xception_tf_dim_ordering_tf_kernels.h5' 
urllib.request.urlretrieve(model_url, model_file)
```

다운 받은 (이미 학습이 완료된)모델을 불러온다.

```python
model = semantic_segmentation()
model.load_pascalvoc_model(model_file) 
```

모델에 이미지를(정확히는 이미지 경로를) 입력하여 segvalues(이미지정보 및 분리정보)와 output(이미지에 대한 sementic segmentation 결과)를 받는다.

```python
segvalues, output = model.segmentAsPascalvoc(img_path) 
```

```python
LABEL_NAMES = [
    'background', 'aeroplane', 'bicycle', 'bird', 'boat', 'bottle', 'bus',
    'car', 'cat', 'chair', 'cow', 'diningtable', 'dog', 'horse', 'motorbike',
    'person', 'pottedplant', 'sheep', 'sofa', 'train', 'tv'
]
```

```python
plt.imshow(output)
plt.show()
```

```python
# 입력한 이미지에 어떤 객체들이 존재하는지 확인한다.
for class_id in segvalues['class_ids']:
    print(LABEL_NAMES[class_id])
```

모델을 통해 확인한 결과, output 이미지에 나온 것처럼 두 종류의 형상으로 이미지가 분리되었으며, 이는 각각 배경과 사람임을 알 수 있었다.  
이미지를 형상별로 분리한 결과를 토대로 사람과 배경을 나누고 배경을 흐리게 만든 뒤, 배경에 다시 사람을 붙여 아웃포커싱 효과를 내보기로 하였다.

```python
# 각 레이블에 적용할 colormap 만들기
# 이미지의 형상들을 sementic 별로 나타내기 위해 색상들을 정하는 작업
# 우리 학습에서는 크게 중요하지 않은 내용이니 너무 신경쓰지 않아도 된다.
colormap = np.zeros((256, 3), dtype = int)
ind = np.arange(256, dtype=int)

for shift in reversed(range(8)):
    for channel in range(3):
        colormap[:, channel] |= ((ind >> channel) & 1) << shift
    ind >>= 3

colormap[:20] #생성한 20개의 컬러맵 출력
```

```python
# LABEL_NAMES에서 person은 15번이기 때문에 output의 사람 이미지 16번째(0부터 시작하므로)
# 색으로 칠해져 있다. BGR 표시이므로, RGB 바꿔야 함
seg_color = (128,128,192)
```

이미지에서 사람 부분을 마스크로 만든다.

```python
# np.all은 지정한 값들이 True인지 여부를 확인하여 nparray로 넘겨준다.
# 일종의 mask 역할을 한다고 보면 될 듯.
# output==seg_color인 값들을 True, 나머지를 False 두어 seg_map의 차원을 축소한다.
seg_map = np.all(output==seg_color, axis=-1) 
print(seg_map.shape) 
# 그레이스케일로 출력, matplotlib에서 임의의 값으로 색을 표현하기 때문에 cmap='gray'를 한 듯 한다.
plt.imshow(seg_map, cmap='gray') #
plt.show()
```

원본 이미지에 mask를 씌워 mask가 배경과 사람을 각각 잘 대표하는지 확인한다.

```python
img_show = img_orig.copy()

# 정수로 타입 변환: seg_map은 boolean원소만을 가지고 있으며
# 흰 부분이 1 이므로 0과 255의 값(검은색과 흰색)만을 가지게 된다.
img_mask = seg_map.astype(np.uint8) * 255

# applyColorMap은 특정 색상 배열(cv2.COLORMAP_JET)을 img_mask에 입힌다.
# 복잡하게 생각할 필요 없이 임의의 색을 img_mask에 입혔다고 생각하면 된다.
color_mask = cv2.applyColorMap(img_mask, cv2.COLORMAP_JET)

# img_show와 color_mask를 섞는다.(8:2의 비율로)
img_show = cv2.addWeighted(img_show, 0.8, color_mask, 0.2, 0.0)

plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show()
```

mask를 확인하였으니, 이제 배경을 blur 처리한다.

```python
# cv2.blur()는 이미지를 흐릿하게 만드는 효과를 지니는데
# 이미 배경이 멀리 있어 많이 흐릿하게 할 필요가 없을 것 같아서
# 필터를 (5,5)정도의 작은 크기로 넣어주었다.(크면 더 흐려짐)
img_orig_blur = cv2.blur(img_orig, (5,5))

# 원본은 당연히, 그리고 여전히 BGR로 되어 있으니 RGB로 바꿔줘야된다.
plt.imshow(cv2.cvtColor(img_orig_blur, cv2.COLOR_BGR2RGB))
plt.show()
```

blur 처리한 배경에서 사람 이미지만 검게 만든 다음, 검은 영역에 원본 사람 이미지를 붙인다.

```python
# cv2.cvtColor(입력 이미지, 색상 변환 코드): 입력 이미지의 색상 채널을 변경
# cv2.COLOR_BGR2RGB: 원본이 BGR 순서로 픽셀을 읽다보니
# 이미지 색상 채널을 변경해야함 (BGR 형식을 RGB 형식으로 변경) 

# 실제 색상이 달라지지는 않지만 현재 img_mask의 데이터형은
# RGB채널이 없는 2차원 ndarray기 때문에 자료형을 변경해 주기 위해
# cvtColor()사용
img_mask_color = cv2.cvtColor(img_mask, cv2.COLOR_GRAY2BGR)

# cv2.bitwise_not(): 이미지가 반전됩니다. 배경이 0 사람이 255 였으나
# 연산을 하고 나면 배경은 255 사람은 0입니다.
img_bg_mask = cv2.bitwise_not(img_mask_color)

# cv2.bitwise_and()을 사용하면 배경만 있는 영상을 얻을 수 있습니다.
# 0과 어떤 수를 bitwise_and 연산을 해도 0이 되기 때문에 
# 사람이 0인 경우에는 사람이 있던 모든 픽셀이 0이 됩니다. 결국 사람이 사라지고 배경만 남아요!
img_bg_blur = cv2.bitwise_and(img_orig_blur, img_bg_mask)
plt.imshow(cv2.cvtColor(img_bg_blur, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# np.where(조건, 참일때, 거짓일때)
# 세그멘테이션 마스크가 255인 부분만 원본 이미지 값을 가지고 오고 
# 아닌 영역은 블러된 이미지 값을 사용합니다.
img_concat = np.where(img_mask_color==255, img_orig, img_bg_blur)
plt.imshow(cv2.cvtColor(img_concat, cv2.COLOR_BGR2RGB))
plt.show()
```

### 2. 인물사진 문제점
첫번째와 두번째 이미지에서는 큰 문제를 찾기 어려웠지만, 세번째 무야호~ 에서는 몇가지 문제점을 찾을 수 있었다.  
가장 눈에 띄는 문제점은 인물이 아닌데도 사람이라고 인식한 부분이 있다는 것이다. 할아버지 사진을 다루던 중 model.segmentAsPascalvoc의 출력값에서부터 할아버지의 왼쪽에 정체불명의 점을 사람이라고 인식하는 것이다. 도통 어떤 부분이 인식되었길래 그런지 정확히는 알 수 없었지만, 명백히 사람이 아닌 부분을 인식한 것만은 확실하다.  
두번째로, 손가락을 제대로 인식하지 못했다. 무야호 할아머지의 오른손(그림에서는 왼쪽에 있는 손)이 부자연스러운 것을 알 수 있는데 다시 원본사진을 확인해 보면 약지와 소지(4,5번째 손가락)이 잘린 것을 알 수 있다.  
마지막으로, 윤곽이 조금씩 잘리거나 작은 공간에 대해서는 제대로 인식하지 못한다. 무야호 할아버지의 오른손(그림에서는 왼쪽에 있는 손)을 보라, 왼손 아래와 어깨, 목 사이에 있는 배경은 확실히 들어갔지만 할아버지의 얼굴(귀)와 손 사이의 공간은 원래 배경색이 그대로 자리잡고 있다는 것을 알 수 있다.(model.segmentAsPascalvoc의 output에서도 제대로 인식하지 못했음) 결국 작은 공간을 제대로 인식할 정도로 정교하지는 못하다는 것이다. 또한 왼손(그림에서는 오른쪽에 있는 손)을 보면 윤곽선을 제대로 잘라내지 못해 편집 전 배경색이 그대로 드러나 있는 것도 볼 수 있다.  
  

결론적으로, 이 모델은 디테일한 부분까지 모두 잡아주지는 못한다. 특히 윤곽선에 의외로 약한 모습을 보이는 것 같다. 윤곽이 제대로 잡혔으면 손가락이 날아가거나 좁은 공간을 놓치지는 않았을 텐데... 그 부분이 아쉽다.

### 3. 해결 방법 말해보기 및 회고
2번에서 언급한 문제점들 중 노드에 힌트가 나와있는 것 위주로 설명하겠다.  
가장 쉽게 이야기할 수 있는 문제는 윤곽이 조금씩 잘리거나 작은 공간을 인식하지 못하고 이를 둘러싼 사물이라고 인식한다는 점이다. 이러한 문제가 생기는 가장 근본적인 이유는 원근감의 인위적인 표현이다. 카메라로 찍는 사진은 촬영하고자 하는 대상에 초점을 맞추는데 초점을 맞춘 대상 외의 사물들은 초점이 맞지 않아 자연스럽게 흐리게 보여 자동으로 아웃포커싱이 된다. 하지만 우리는 특정 대상을 AI로 식별한 다음 대상 외의 나머지 사물들을 인위적으로 흐리게 만들었기 때문에 대상 식별에 문제가 있을 경우 사진의 거리감도 이상하게 표현되는 것이다.  
  
이에 대한 해결 방안으로 노드에서는 3D이미지 센서를 언급하고 있다. 아래의 설명은 구글에 '3D 이미지 센서란'으로 검색시 가장 상위에 나오는 검색 결과의 일부다.  

'3D 센싱은 목표물에 발사된 빛의 깊이 정보와 각도 분해능 기능을 통해 거리를 파악하고, 물체를 인식하는 기술이다.'   

모든 사람들이 2차원의 평면을 눈으로 보지만 눈으로 본 장면에 대해 모종의 과정을 거쳐 3차원으로 이해하는 것처럼 3D 센서도 이미지를 3차원의 형태로 인식할 수 있다는 것이다. 2차원 이미지 안에서도 원근을 인식하여 3차원으로 이해 할 수 있다면 사물의 원근에 따른 시각적 차이를 더욱 세부적으로 분간할 수 있고 위와 같은 문제를 해결하는데 도움이 될 것으로 보인다.  
  
세번째 합성사진인 무야호 할아버지 사진만 보아도 알 수 있다. 할아버지 손 사이의 작은 공간이 할아버지의 일부로 인식된 데에는 할아버지의 얼굴을 3차원적으로 인식하는 것이 불가했기 때문에 사이의 작은 공간을 잡아내지 못하고 전체를 사람으로 인식하였을 것이다. 하지만 3D 센서라면 할아버지의 얼굴과 그 사이의 배경은 물론 야호를 외치는 손까지 원근을 세밀하게 파악할 수 있기 때문에 이러한 문제는 발생하지 않을 것이다.  
  
지금까지 이미지를 인식하고 편집하여 아웃포커싱 사진 및 배경전환 크로마키 사진을 만들고 문제점을 찾아본 다음 이에 대한 해결책도 확인해 보았다. 과제 자체는 결과가 눈에 보이기도 하고 하나씩 조작해보는 재미가 있어서 그런지 즐겁게 했지만 한편으로는 조금 고민이 되는 부분도 있었다.  
  
'뭔가 정말 좋고 성과를 낼 수 있는 모델들은 다 가져와서 쓰는 것 같은데, 그렇다면 AI 공부를 깊게 하는 것이 어떤 의미를 가질 수 있을까?'  
  
이번 과제만 해도 AI에 사용되는 모델은 가져온 것이다. 이미 많은 이들이 사용해보고 일정 수준 최적화까지 진행된 완성된 모델을 사용한 것이다. 그렇게 모델을 가져와 사용하는 파트는 빠르게 지나가고 대부분은 파이썬을 가지고 포토샵을 한 것 같은 느낌이다. 어느 정도 이론적으로 배우고, 과제를 따라하다 보니 CV니 NLP나 이러한 패턴이 반복된다는 생각을 지울 수가 없다. 그렇다면 다시 웹이나 앱 같은 어플리케이션 프로그래밍에 집중해야 하는 것인가??? 하는 생각이 들기 시작한 것이다. 프로젝트(아이펠톤?)에 대한 이야기가 조금씩 나오는 상황이다 보니 더욱 그렇다. 'AI교육만 가지고 프로젝트를 완료하는 것이 쉽지는 않을 것 같은데 그렇다면 결국...?'이라는 생각이 게속 들게 되는 것이다. AI를 배우고 우리가 현업에서 사용한다면 당연히 개발적인 관점에서의 코딩도 어느 정도 하게 될 것으로 보이는데 여기에 대해서는 준비하고 있지 않다.  
생각이 많아지는 일요일 밤이다.
